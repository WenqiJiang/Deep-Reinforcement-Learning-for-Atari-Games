```python
import pandas as pd
import json
import numpy as np
with open('./vgg_sarsa_BreakoutDeterministic-v4_log.json') as f:
    data = json.load(f)
print (type(data))
#print (pd.DataFrame(data))
df = pd.DataFrame(data)
print(df.shape)
display(df.head(10))
df.describe()
```

```python
loss_hist = df["loss"].tolist()
mse_hist = df["mean_absolute_error"].tolist()
mq_hist = df["mean_q"].tolist()
meps_hist = df["mean_eps"].tolist()
er_hist = df["episode_reward"].tolist()
```

```python
import matplotlib.pyplot as plt
N=200
loss_hist = np.convolve(loss_hist, np.ones((N,))/N, mode='valid')
x = np.linspace(0,17499, len(loss_hist))
plt.plot(x, loss_hist, label="loss")
plt.legend()
plt.show()
```

```python
N=200
mse_hist = np.convolve(mse_hist, np.ones((N,))/N, mode='valid')
x = np.linspace(0,17499, len(mse_hist))
plt.plot(x, mse_hist, label="mse")
plt.legend()
plt.show()
```

```python
N=200
mq_hist = np.convolve(mq_hist, np.ones((N,))/N, mode='valid')
x = np.linspace(0,17499, len(mq_hist))
plt.plot(x, mq_hist, label="mean_q")
plt.legend()
plt.show()
```

```python
x = np.linspace(0,17499, len(meps_hist))
plt.plot(x, meps_hist, label="mean_eps")
plt.legend()
plt.show()
```
