```python
import pandas as pd
import json
import numpy as np
```

# DQN

```python
with open('dqn_BreakoutDeterministic-v4_log.json') as f:    
    data_dqn = json.load(f)

df_dqn = pd.DataFrame(data_dqn)

loss_hist_dqn = df_dqn["loss"].tolist()
mae_hist_dqn = df_dqn["mean_absolute_error"].tolist()
mq_hist_dqn = df_dqn["mean_q"].tolist()
meps_hist_dqn = df_dqn["mean_eps"].tolist()
er_hist_dqn = df_dqn["episode_reward"].tolist()

N=100
loss_hist_dqn_sm = np.convolve(loss_hist_dqn, np.ones((N,))/N, mode='valid')
mae_hist_dqn_sm =  np.convolve(mae_hist_dqn, np.ones((N,))/N, mode='valid')
mq_hist_dqn_sm =  np.convolve(mq_hist_dqn, np.ones((N,))/N, mode='valid')
meps_hist_dqn_sm =  np.convolve(meps_hist_dqn, np.ones((N,))/N, mode='valid')
er_hist_dqn_sm =  np.convolve(er_hist_dqn, np.ones((N,))/N, mode='valid')


import matplotlib.pyplot as plt
fig, axes = plt.subplots(1,5,figsize=(18,4))
fig.suptitle('DQN')

axes[0].plot(loss_hist_dqn, label="loss")
axes[0].plot(loss_hist_dqn_sm, label="loss_smooth")
axes[0].legend()
axes[0].set_xlabel('episode')
axes[0].set_ylabel('metric')

axes[1].plot(mae_hist_dqn, label="mean_absolute_error")
axes[1].legend()
axes[1].set_xlabel('episode')

axes[2].plot(mq_hist_dqn, label="mean_q")
axes[2].legend()
axes[2].set_xlabel('episode')

axes[3].plot(meps_hist_dqn, label="mean_eps")
axes[3].legend()
axes[3].set_xlabel('episode')

axes[4].plot(er_hist_dqn, label="episode_reward")
axes[4].plot(er_hist_dqn_sm, label="er_smooth")
axes[4].legend()
axes[4].set_xlabel('episode')

plt.show()
```
